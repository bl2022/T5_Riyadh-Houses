```python
#This method to show outliers in box plot.
sns.boxplot(data=houses)
```

```python
def clean_outlier(column):
    q1 = np.quantile(houses[column], 0.25)
    q3 = np.quantile(houses[column], 0.75)
    IQR = q3 - q1
    
    min_range = q1 - 1.5 * IQR
    max_range = q3 + 1.5 * IQR
    
    mean = houses[column].mean()
    houses.loc[(houses[column] < min_range), column] = mean
    houses.loc[(houses[column] > max_range), column] = mean
    

clean_outlier('size')
clean_outlier('price')
clean_outlier('bedrooms')
clean_outlier('bathrooms')
```

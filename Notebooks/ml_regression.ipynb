## Regression

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
lm = LinearRegression()
```

```python
X = pd.DataFrame()
X['size'] = last_houses['size']
X = pd.concat([X, last_houses.iloc[:, 6:]], axis=1)
X.head(3)
```

```python
y = last_houses['price']
y.head(3)
```

```python
def rmse(y_actual, y_predicted):
    return sqrt(mean_squared_error(y_actual, y_predicted))

def rmse_error(reg_model, x, y_actual):
    y_predicted = reg_model.predict(x)
    return rmse(y_actual, y_predicted)
```

```python
#this function is train and test tha data and evaluate the errors for linear regression.
def learn_linear(X, y, results, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
    
    scaler = StandardScaler()
    
    X_train_scaled = scaler.fit_transform(X_train.values)
    X_test_scaled = scaler.transform(X_test.values)
    
    lm = LinearRegression()
    lm.fit(X_train, y_train)
    
    train_r2 = lm.score(X_train, y_train)
    test_r2 = lm.score(X_test, y_test)
    
    rmse_error_train = rmse_error(lm, X_train, y_train)
    rmse_error_test = rmse_error(lm, X_test, y_test)
    
    results['train_r2'].append(train_r2)
    results['test_r2'].append(test_r2)
    results['rmse_error_train'].append(rmse_error_train)
    results['rmse_error_test'].append(rmse_error_test)
    
    return results
```

```python
#this function is train and test tha data and evaluate the errors for ridge regression.
def learn_ridge(X, y, results, alpha, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
    
    scaler = StandardScaler()
    
    X_train_scaled = scaler.fit_transform(X_train.values)
    X_test_scaled = scaler.transform(X_test.values)
    
    lm_ridge = Ridge(alpha=alpha)
    lm_ridge.fit(X_train_scaled, y_train)
    
    train_r2 = lm_ridge.score(X_train_scaled, y_train)
    test_r2 = lm_ridge.score(X_test_scaled, y_test)
    
    rmse_error_train = rmse_error(lm_ridge, X_train_scaled, y_train)
    rmse_error_test = rmse_error(lm_ridge, X_test_scaled, y_test)
    
    results['train_r2'].append(train_r2)
    results['test_r2'].append(test_r2)
    results['rmse_error_train'].append(rmse_error_train)
    results['rmse_error_test'].append(rmse_error_test)
    
    return results
```

```python
#this function is train and test tha data and evaluate the errors for polynomial regression.
def learn_poly(X, y, results, degree, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
    
    poly = PolynomialFeatures(degree=degree) 
    
    X_train_poly = poly.fit_transform(X_train.values)
    X_test_poly = poly.transform(X_test.values)

    lm_poly = LinearRegression()
    
    lm_poly.fit(X_train_poly, y_train)
    
    train_r2 = lm_poly.score(X_train_poly, y_train)
    test_r2 = lm_poly.score(X_test_poly, y_test)
    
    rmse_error_train = rmse_error(lm_poly, X_train_poly, y_train)
    rmse_error_test = rmse_error(lm_poly, X_test_poly, y_test)
    
    results['train_r2'].append(train_r2)
    results['test_r2'].append(test_r2)
    results['rmse_error_train'].append(rmse_error_train)
    results['rmse_error_test'].append(rmse_error_test)
    
    return results
```

```python
results = {'train_r2': [], 'test_r2': [], 'rmse_error_train': [], 'rmse_error_test': []}

results = learn_linear(X, y, results)
results = learn_ridge(X, y, results, alpha=1)
results = learn_ridge(X, y, results, alpha=2)
results = learn_ridge(X, y, results, alpha=3)
```

```python
results_df = pd.DataFrame(results,
                          index=['Linear Model', 'Ridge Model Alpha=1', 'Ridge Model Alpha=2', 'Ridge Model Alpha=3'])


results_df
```
